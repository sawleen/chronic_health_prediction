## Feature Engineering

```python
# Count the number of reports per member
df_by_mem = df.groupby('mem_id')['lab_report_no'].count().reset_index(name="CountLabReport")
```

## Read in target dataset (members)

```python
TGTFILE = "member_train_data_final.pkl"

# If csv format
#df_tgt = pd.read_csv(TGTFILE)

# If pkl format
with open(TGTFILE, 'rb') as pklfile:
    df_tgt = pickle.load(pklfile)
```

```python
df_tgt.head()
```

```python
# Merge the report numbers column to the Members dataset
# Replace those members with no lab test to be 0

df_output = pd.merge(df_tgt, df_by_mem, how='left', on='mem_id')
df_output['CountLabReport'] = df_output['CountLabReport'].fillna(value=0).astype('int64')
```

```python
df_output.head()
```

```python
# Check that the lab report count is well merged into the new members dataset
df_output['CountLabReport'].value_counts()
```

## Merge categorized lab report to members dataset

```python
# Create a subset of dataframe with columns starting with CAT*
CAT = [col for col in df if col.startswith("CAT")]
memCAT = ["mem_id","requested_date"] + CAT
df[memCAT]
```

```python
df_lab = df[memCAT].replace('not tested', '')
```

```python
df_lab
```

```python
df_lab2=df_lab.sort_values(by = "requested_date")
```

```python
df_lab2
```

```python
# Step through the rows, if the mem_id is a repeated case,
# decide whether to replace all with the new values (if av)
# but if new values is null, then replace them with old values (better than null)

for i, x in enumerate(df_lab2['mem_id']):
    #print(i)
    for j in range(i):
        #print(i,j)
        if(df_lab2['mem_id'][i] == df_lab2['mem_id'][j]):
            #print(j)
            print("row",i,": ", df_lab2['mem_id'][i], 'vs',df_lab2['mem_id'][j])
            # Check for each column CAT*
            for k, z in enumerate(CAT):
                if(df_lab2[z][i] != ''):
                    # replace the old value with the latest value
                    df_lab2[z][j] = df_lab2[z][i]
                else:
                    # replace the new (null) value with the old value
                    df_lab2[z][i] = df_lab2[z][j]

    
```

```python
df_lab2 = df_lab2.drop('requested_date', axis=1)
```

```python
# double check that repeated mem_id all have consistent latest lab results
df_lab2.duplicated().sum()
#df_lab2.duplicated(subset=CAT+["mem_id"]).sum()
```

```python
# remove the duplicates accordingly
df_lab2 = df_lab2.drop_duplicates()
#df_lab2 = df_lab2.drop_duplicates(subset=CAT+["mem_id"])
```

```python
df_lab2
```

```python
# Merge the lab categorical column to the Members dataset

df_output2 = pd.merge(df_output, df_lab2, how='left', on='mem_id')
```

```python
# Interim saving of df to csv format
df_output2.to_csv('member_train_data_final_v2_CAT.csv', index=False, header=True)
# Interim saving of df to pkl format
df_output2.to_pickle("member_train_data_final_v2_CAT.pkl")
```

### Scan for number of lab reports

```python
df_booking.insert(loc=len(df_booking.columns),column='CountLabReport',value=0)
```

```python
df_booking.head()
```

```python
df_booking.info(verbose=True, show_counts=True)
```

```python
# For each booking id, note the member id, then
# scan through the lab report for the same member,
# if there is a laboratory test requested by the same member
# prior to the consultation date, add +1 to the CountLabReport

for i, x in enumerate(df_booking['mem_id']):
    for j, y in enumerate(df['mem_id']):
        if(x == y):
            if(df_booking['booking_timestamp'][i] > df['requested_date'][j]):
                print(x,'lab was done previously')
                df_booking['CountLabReport'][i] +=1 
    
```

```python
# Check the list of bookings that have non-zero CountLabReport
df_booking[df_booking['CountLabReport'] > 0]
```
