```python
df['CAT|LIPID PROFILE|HDL Cholesterol|mg/dL']=pd.cut(df['LIPID PROFILE|HDL Cholesterol|mg/dL'],bins, labels=names, ordered=False)
df['CAT|LIPID PROFILE|HDL Cholesterol|mg/dL']=df['CAT|LIPID PROFILE|HDL Cholesterol|mg/dL'].cat.add_categories("not tested")
# substitute all NaN with "not tested"
df['CAT|LIPID PROFILE|HDL Cholesterol|mg/dL'].fillna('not tested', inplace=True)
```

```python
#          low    normal      high
# Male     <0.7   0.7-1.3     >1.3
# Female   <0.6   0.6-1.1     >1.1
# Unknown  <0.65  0.65-1.2    >1.2 (mid values of F and M)

df.insert(loc=len(df.columns),column='CAT|SERUM CREATININE|Creatinine|mg/dL', value='')
```

```python
# Mapping to the respective categories based on gender

for i, x in enumerate(df['SERUM CREATININE|Creatinine|mg/dL']):
    if df['gender'][i] == "Male":
        if x < 0.7:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="low"
        elif (x < 1.3) & (x >= 0.7):
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="normal"
        elif x >= 1.3:
             df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="high"   
        else:
             df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="not tested"   
    elif df['gender'][i] == "Female":
        if x < 0.6:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="low"
        elif (x < 1.1) & (x >= 0.6):
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="normal"
        elif x >= 1.1:
             df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="high"   
        else:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL'] ="not tested"   
    elif df['gender'][i] == "Unknown":
         if x < 0.65:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="low"
         elif (x < 1.2) & (x >= 0.65):
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="normal"
         elif x >= 1.2:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="high"   
         else:
            df.loc[i,'CAT|SERUM CREATININE|Creatinine|mg/dL']="not tested"   
    else:
         print("There is Unknown here in:", i)
```
